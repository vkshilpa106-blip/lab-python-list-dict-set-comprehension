````python
def initialize_inventory(products):
    """1. Initializes the inventory dictionary using an explicit loop for safe Jupyter inputs."""
    inventory = {}
    print("\n--- Initialize Inventory ---")
    for product in products:
        qty_input = input(f"Enter the quantity of {product}s available: ").strip()
        inventory[product] = int(qty_input)
    return inventory


def get_customer_orders():
    """2. Prompts user for order count and collects names cleanly using a loop."""
    print("\n--- Get Customer Orders ---")
    num_orders_input = input("Enter the number of customer orders: ").strip()
    num_orders = int(num_orders_input)
    
    customer_orders = set()
    for _ in range(num_orders):
        order = input("Enter the name of a product that a customer wants to order: ").strip()
        if order:
            customer_orders.add(order)
            
    return customer_orders


def update_inventory(customer_orders, inventory):
    """4. Updates counts and removes items with 0 stock using a dictionary comprehension."""
    # Comprehension used here to safely create a lowercase reference map
    inventory_lowercase = {product.lower(): product for product in inventory}

    # Deduct stock for matching orders
    for order in customer_orders:
        order_lower = order.lower()
        if order_lower in inventory_lowercase:
            original_name = inventory_lowercase[order_lower]
            if inventory[original_name] > 0:
                inventory[original_name] -= 1

    # Dictionary Comprehension: Filters out any keys whose quantity dropped to 0
    updated_inventory = {prod: qty for prod, qty in inventory.items() if qty > 0}
    
    # Apply changes directly to our main inventory reference
    inventory.clear()
    inventory.update(updated_inventory)


def calculate_order_statistics(customer_orders, products):
    """Calculates order statistics using a list filter operation."""
    total_products_ordered = len(customer_orders)
    products_lower = [p.lower() for p in products]
    
    # Counts valid products matching the store catalog
    valid_orders_count = sum(1 for order in customer_orders if order.lower() in products_lower)
    
    percentage_ordered = (valid_orders_count / len(products)) * 100 if products else 0.0
    return total_products_ordered, percentage_ordered


def print_order_statistics(order_statistics):
    """Prints the calculated order statistics."""
    total, percentage = order_statistics
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total}")
    print(f"Percentage of Unique Products Ordered: {percentage:.1f}")


def print_updated_inventory(inventory):
    """6. Prints the updated inventory layout."""
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")


def calculate_total_price(customer_orders):
    """3 & 5. Prompts for product prices sequentially and prints the total sum."""
    print()  # Inserts a formatting blank line before prompt sequences begin
    total_price = 0.0
    
    for product in customer_orders:
        price_input = input(f"Enter the price of {product}: ").strip()
        total_price += float(price_input)
        
    print(f"Total Price: {total_price:.1f}")
    return total_price


# 7. Appropriate execution sequence setup
if __name__ == "__main__":
    # Base configuration array matches assignment requirements
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    
    # Execution pipeline steps
    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders()
    
    # Compile metrics before 0-stock products are removed from memory tracking
    order_statistics = calculate_order_statistics(customer_orders, products)
    update_inventory(customer_orders, inventory)
    
    # Format and terminal print results block
    print_order_statistics(order_statistics)
    print_updated_inventory(inventory)
    calculate_total_price(customer_orders)

````
